# Example 1: Calculate power of single inflow and plot

## Pramble

```python
import floris
from floris import *
# from floris import floris
from floris import FlorisModel
import matplotlib.pyplot as plt
```

## Inputs

```python
x = [0, 500, 1000]
y = [0,0,0]
wd = [270]
ws = [8]
ti = [0.06]
```

```python
fmodel = FlorisModel('/home/lavender/Studies/Design of Wind Farms/Tutorial/Week3/gch.yaml')
fmodel.set(layout_x=x, layout_y=y,wind_directions=wd,wind_speeds=ws,turbulence_intensities=ti)
```

```python
fmodel.run()
```

```python
turbine_powers=fmodel.get_turbine_powers()
farm_power = fmodel.get_farm_power()
print("Turbine Power (MW): ", turbine_powers)
print("Farm Power (MW): ", farm_power)
```

```python
## Plot flow field

### Create Figure

fig,ax = plt.subplots()

### Calculate Plane

horizontal_plane = fmodel.calculate_horizontal_plane(height=90, x_resolution=200, y_resolution=100)

### Plot

# visualise_cut_plane()

visualize_cut_plane(horizontal_plane, ax=ax, color_bar=True, title="Simple Flow Field Visualization")
```
