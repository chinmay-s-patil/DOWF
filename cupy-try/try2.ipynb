```python
import time
import cupy as cp

cp.cuda.Device(0).use()   # RX 6600M

n = 24
a = cp.random.random((n, n), dtype=cp.float32)
b = cp.random.random((n, n), dtype=cp.float32)

cp.cuda.Stream.null.synchronize()
t0 = time.perf_counter()
c = a @ b
cp.cuda.Stream.null.synchronize()
t1 = time.perf_counter()

print("GPU time:", t1 - t0)
print("device:", cp.cuda.runtime.getDeviceProperties(0)["name"].decode())
```
