```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
U_inf = 7.5        # freestream wind speed (m/s)
Ct = 0.766
kw = 0.075
D = 130.0

csv_file = "/home/lavender/Studies/Design of Wind Farms/Assignments/Assignment2/IEA_Reference_3.4MW_130.csv"   # download from GitHub
```

```python
data = pd.read_csv(csv_file)

wind_speeds = data["Wind Speed [m/s]"].values
powers = data["Power [kW]"].values

def power_from_curve(U):
    """Interpolate power curve"""
    return np.interp(U, wind_speeds, powers)
```

```python
def jensen_deficit(x):
    """Velocity deficit from one turbine at distance x"""
    return (1 - np.sqrt(1 - Ct)) / (1 + 2*kw*x/D)**2

# distances between turbines (in meters)
x12 = 5*D
x23 = 5*D
x13 = 10*D

# compute individual deficits
d12 = jensen_deficit(x12)
d23 = jensen_deficit(x23)
d13 = jensen_deficit(x13)
```

```python
# Turbine 1 sees free wind
U1 = U_inf

# Turbine 2 sees wake from T1
U2 = U_inf * (1 - d12)

# Turbine 3 sees wakes from T1 and T2 (RSS superposition)
d_total_T3 = np.sqrt(d13**2 + d23**2)
U3 = U_inf * (1 - d_total_T3)

# compute powers
P1_0 = power_from_curve(U1)
P2_0 = power_from_curve(U2)
P3_0 = power_from_curve(U3)

# relative powers
rel_0 = np.array([P1_0, P2_0, P3_0]) / P1_0
```

```python
U1_270 = U2_270 = U3_270 = U_inf

P1_270 = power_from_curve(U1_270)
P2_270 = power_from_curve(U2_270)
P3_270 = power_from_curve(U3_270)

rel_270 = np.array([P1_270, P2_270, P3_270]) / P1_270
```

```python
print("\n===== WIND DIRECTION 0° (WAKE CASE) =====")
print(f"Turbine wind speeds: {U1:.3f}, {U2:.3f}, {U3:.3f} m/s")

print("\nAbsolute Power [kW]")
print(f"T1: {P1_0:.1f}")
print(f"T2: {P2_0:.1f}")
print(f"T3: {P3_0:.1f}")

print("\nRelative Power Pn/P1")
print(f"T1: {rel_0[0]:.3f}")
print(f"T2: {rel_0[1]:.3f}")
print(f"T3: {rel_0[2]:.3f}")

print("\n===== WIND DIRECTION 270° (NO WAKE) =====")
print("\nAbsolute Power [kW]")
print(f"T1: {P1_270:.1f}")
print(f"T2: {P2_270:.1f}")
print(f"T3: {P3_270:.1f}")

print("\nRelative Power Pn/P1")
print(f"T1: {rel_270[0]:.3f}")
print(f"T2: {rel_270[1]:.3f}")
print(f"T3: {rel_270[2]:.3f}")
```
